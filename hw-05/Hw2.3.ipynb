```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

file_path = "bestsellers with categories.csv"

df = pd.read_csv(file_path)

print(df.head())
```

```python
print(f"Про скільки книг зберігає дані датасет? Відповідь: {df.shape[0]}")
```

```python
df.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']
```

```python
# Перевірте, чи у всіх рядків вистачає даних: виведіть кількість пропусків (na) у кожному зі стовпців (використовуйте функції isna та sum)

missing_values = df.isna().sum()

print(missing_values)
```

```python
# Чи є в якихось змінних пропуски? (Так / ні)
# Відповідь: ні
```

```python
# Перевірте, які є унікальні значення в колонці genre (використовуйте функцію unique)

unique_genres = df["genre"].unique()

print(f"Які є унікальні жанри? Відповідь: {unique_genres[0]}, {unique_genres[1]}")
```

```python
# Тепер подивіться на розподіл цін: побудуйте діаграму (використовуйте kind='hist')

df["price"].hist(bins=20, edgecolor="black")

plt.xlabel("Ціна")
plt.ylabel("Кількість")
plt.title("Розподіл цін")

plt.show()
```

```python
# Визначте, яка ціна у нас максимальна, мінімальна, середня, медіанна (використовуйте функції max, min, mean, median)

max_price = df["price"].max()

min_price = df["price"].min()

mean_price = df["price"].mean()

median_price = df["price"].median()

print(f"Максимальна ціна? Відповідь: {max_price}")
print(f"Мінімальна ціна? Відповідь: {min_price}")
print(f"Середня ціна? Відповідь: {mean_price}")
print(f"Медіанна ціна? Відповідь: {median_price}")
```

```python
max_rating = df["user_rating"].max()

print(f"Який рейтинг у датасеті найвищий? Відповідь: {max_rating}")
```

```python
count_max_rating = df[df["user_rating"] == max_rating].shape[0]

print(f"Скільки книг мають такий рейтинг? Відповідь: {count_max_rating}")
```

```python
max_reviews = df["reviews"].max()

most_reviewed_books = df[df["reviews"] == max_reviews]

print(f"Яка книга має найбільше відгуків? Відповідь: {most_reviewed_books["name"].values[0]}")
```

```python
top_50_2015 = df[(df["year"] == 2015)]

most_expensive_book = top_50_2015.loc[top_50_2015["price"].idxmax(), "name"]

print(f"З тих книг, що потрапили до Топ-50 у 2015 році, яка книга найдорожча (можна використати проміжний датафрейм)? Відповідь: {most_expensive_book}")
```

```python
fiction_top_50_2010 = df[(df["year"] == 2010) & (df["genre"] == "Fiction")]

count_fiction_top_50_2010 = fiction_top_50_2010.shape[0]

print(f"Скільки книг жанру Fiction потрапили до Топ-50 у 2010 році (використовуйте &)? Відповідь: {count_fiction_top_50_2010} ")
```

```python
high_rating_books = df[(df["year"].isin([2010, 2011])) & (df["user_rating"] == 4.9)]

count_high_rating_books = high_rating_books.shape[0]

print(f"Скільки книг з рейтингом 4.9 потрапило до рейтингу у 2010 та 2011 роках (використовуйте | або функцію isin)? Відповідь: {count_high_rating_books} ")
```

```python
cheap_books_2015 = df[(df["year"] == 2015) & (df["price"] < 8)]

sorted_cheap_books_2015 = cheap_books_2015.sort_values(by="price")

print(sorted_cheap_books_2015)
```

```python
last_book = sorted_cheap_books_2015.iloc[-1]["name"]
print(f"Яка книга остання у відсортованому списку? Відповідь: {last_book}")
```

```python
price_stats = df.groupby("genre")["price"].agg(["min", "max"])
fiction_max_price = price_stats.loc["Fiction", "max"]
fiction_min_price = price_stats.loc["Fiction", "min"]
non_fiction_max_price = price_stats.loc["Non Fiction", "max"]
non_fiction_min_price = price_stats.loc["Non Fiction", "min"]

print(f"Максимальна ціна жанру Fiction? Відповідь: {fiction_max_price}")
print(f"Мінімальна ціна жанру Fiction? Відповідь: {fiction_min_price}")
print(f"Максимальна ціна жанру Non Fiction? Відповідь: {non_fiction_max_price}")
print(f"Мінімальна ціна жанру Non Fiction? Відповідь: {non_fiction_min_price}")
```

```python
author_book_count = df.groupby("author")["name"].count().reset_index()
author_book_count.columns = ["author", "book_count"]

print(f"Якої розмірності вийшла таблиця? Відповідь: {author_book_count.shape}")
```

```python
top_author = author_book_count.loc[author_book_count["book_count"].idxmax()]

print(f"Який автор має найбільше книг? Відповідь: {top_author["author"]}")
print(f"Скільки книг у цього автора? Відповідь: {top_author["book_count"]}")
```

```python
author_avg_rating = df.groupby("author", as_index=False).agg({"user_rating": "mean"})

min_rating_author = author_avg_rating.loc[author_avg_rating["user_rating"].idxmin()]

print(f"У якого автора середній рейтинг мінімальний? Відповідь: {min_rating_author["author"]}")
print(f"Який у цього автора середній рейтинг? Відповідь: {min_rating_author["user_rating"]}")
```

```python
author_stats = pd.concat([author_book_count, author_avg_rating["user_rating"]], axis=1)

print(author_stats)
```

```python
author_stats = author_stats.sort_values(by=["book_count", "user_rating"], ascending=[True, True])

print(author_stats)
```

```python
print(f"Який автор перший у списку? Відповідь: {author_stats.iloc[0]["author"]}")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='price', hue='genre', bins=20, kde=True)
plt.title('Розподіл цін книг за жанрами')
plt.xlabel('Ціна')
plt.ylabel('Кількість книг')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='year', y='user_rating', hue='genre', marker='o')
plt.title('Середні рейтинги книг за роками')
plt.xlabel('Рік')
plt.ylabel('Середній рейтинг')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='genre', y='reviews')
plt.title('Розподіл кількості рецензій за жанрами')
plt.xlabel('Жанр')
plt.ylabel('Кількість рецензій')
plt.yscale('log')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='price', y='reviews', hue='genre')
plt.title('Співвідношення ціни та кількості рецензій')
plt.xlabel('Ціна')
plt.ylabel('Кількість рецензій')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='year', hue='year', legend=False, palette='viridis')
plt.title('Кількість книг за роками')
plt.xlabel('Рік')
plt.ylabel('Кількість книг')
plt.xticks(rotation=45)
plt.show()
```
