```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

file_path = "2017_jun_final - 2017_jun_final.csv"

# 1. Прочитайте файл 2017_jun_final.csv за допомогою методу read_csv

df = pd.read_csv(file_path)
```

```python
# 2. Прочитайте отриману таблицю, використовуючи метод head

print(df.head())
```

```python
# 3. Визначте розмір таблиці за допомогою методу shape

df.shape
```

```python
# 4. Визначте типи всіх стовпців за допомогою dataframe.dtypes

df.dtypes
```

```python
# 5. Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

missing_ratio = df.isnull().sum() / len(df)
print(missing_ratio)
```

```python
# 6. Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"

cols_to_keep = ["Мова програмування"] + df.columns[df.isnull().sum() == 0].tolist()
df_cleaned = df[cols_to_keep]
print(df_cleaned)
```

```python
# 7. Знову порахуйте, яка частка пропусків міститься в кожній колонці і переконайтеся, що залишився тільки стовпець "Мова.програмування"

missing_ratio = df_cleaned.isnull().sum() / len(df_cleaned) * 100
print(missing_ratio)
```

```python
# 8. Видаліть усі рядки у вихідній таблиці за допомогою методу dropna

df_cleaned = df.dropna()
```

```python
# 9. Визначте новий розмір таблиці за допомогою методу shape

df_cleaned.shape
```

```python
# 10. Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python

python_data = df[df["Мова програмування"] == "Python"]

print(python_data)
```

```python
# 11. Визначте розмір таблиці python_data за допомогою методу shape

python_data.shape
```

```python
# 12. Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"

grouped_data = python_data.groupby("Посада")
```

```python
# 13. Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

salary_stats = python_data.groupby("Посада")["Зарплата на місяць"].agg(["min", "max"]).reset_index()

print(salary_stats.head())
```

```python
# 14. Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

salary_avg = python_data.groupby("Посада")["Зарплата на місяць"].mean().to_dict()

def fill_avg_salary(row):
    return salary_avg.get(row["Посада"], None)

python_data.loc[:, "avg"] = python_data.apply(fill_avg_salary, axis=1)

unique_salaries = python_data[["Посада", "avg"]].drop_duplicates()
print(unique_salaries)
```

```python
# 15. Створіть описову статистику за допомогою методу describe для нового стовпчика.

stats = python_data["avg"].describe()
print(stats)
```

```python
# 16. Збережіть отриману таблицю в CSV файл

python_data.to_csv("python_data.csv", index=False, encoding="utf-8")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=python_data, x='Зарплата на місяць', bins=30, kde=True, color='skyblue')
plt.title('Розподіл зарплат спеціалістів Python')
plt.xlabel('Зарплата на місяць')
plt.ylabel('Кількість')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='exp', y='Зарплата на місяць', hue='Розмір компанії')
plt.title('Залежність зарплати від досвіду роботи')
plt.xlabel('Загальний досвід роботи (роки)')
plt.ylabel('Зарплата на місяць')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
specialization_salary = df.groupby('Спеціалізація')['Зарплата на місяць'].mean().sort_values()
specialization_salary.plot(kind='line', marker='o')
plt.title('Середня зарплата по спеціалізаціях')
plt.xlabel('Спеціалізація')
plt.ylabel('Середня зарплата на місяць')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
specialization_counts = df['Спеціалізація'].value_counts()
specialization_counts.plot(kind='bar', color='teal')
plt.title('Кількість програмістів за спеціалізаціями')
plt.xlabel('Спеціалізація')
plt.ylabel('Кількість')
plt.xticks(rotation=45)
plt.show()
```
